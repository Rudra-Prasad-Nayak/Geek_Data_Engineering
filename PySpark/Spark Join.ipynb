```python
import pyspark
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import SparkSession
```

```python
empSchema = StructType([
    StructField("EMPNO", IntegerType(), True),
    StructField("ENAME", StringType(), True),
    StructField("JOB", StringType(), True),
    StructField("MGR", IntegerType(), False),
    StructField("HIREDATE", StringType(), True),
    StructField("SAL", IntegerType(), True),
    StructField("COMM", IntegerType(), False),
    StructField("DEPTNO", IntegerType(), True),
])
```

```python
df_emp = spark.read.format('csv').option('header', True).load('/Volumes/workspace/default/spark/Employees.csv', schema = empSchema)

df_emp = df_emp.withColumn(
    "HIREDATE",
    when(
        year(expr("try_to_date(HIREDATE, 'dd-MMM-yy')")) > 2025,
        add_months(expr("try_to_date(HIREDATE, 'dd-MMM-yy')"), -1200)
    ).otherwise(expr("try_to_date(HIREDATE, 'dd-MMM-yy')"))
)

df_emp.display()
```

```python
deptSchema = StructType([
    StructField("DEPTNO", IntegerType(), True),
    StructField("DNAME", StringType(), True),
    StructField("LOC", StringType(), True),
])
```

```python
df_dept = spark.read.format('csv').option('header', True).load('/Volumes/workspace/default/spark/Departments.csv', schema = deptSchema)
df_dept.display()
```

```python
salgradeSchema = StructType([
    StructField("GRADE", IntegerType(), True),
    StructField("LOSAL", IntegerType(), True),
    StructField("HISAL", IntegerType(), True),
])
```

```python
df_salgrade = spark.read.format('csv').option('header', True).load('/Volumes/workspace/default/spark/Salgrade.csv', schema = salgradeSchema)
df_salgrade.display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'inner').display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'left').display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'right').display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'full_outer').display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'left_anti').display()
```

```python
df_emp.join(df_dept, on = [df_emp['DEPTNO'] == df_dept['DEPTNO']], how = 'left_semi').display()
```

```python
df_emp.join(df_salgrade, on = [(df_emp.SAL >= df_salgrade.LOSAL) & (df_emp.SAL <= df_salgrade.HISAL),], how = 'inner').display()
```

```python
df_emp.join(df_salgrade, on = [df_emp.SAL.between(df_salgrade.LOSAL, df_salgrade.HISAL)], how = 'inner').display()
```
