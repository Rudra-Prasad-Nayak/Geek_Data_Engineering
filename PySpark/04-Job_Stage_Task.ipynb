```python
import re
#regular expression module
```

```python
df_addr = spark.createDataFrame([(1,'Cogniza@nt SOlu😄tions'),(2,'Namo Namo'),(3,'Hotel I😜ndsu*try'),(4, "acknge   sgywx %d dedd👌")],schema = ['id','Addr'])
df_addr.display()
```

```python
@udf
def remove_space(x):
  return re.sub(" ","",x)
```

```python
@udf
def clean_char(x):
  return re.sub("[@%&*(^)]","",x)
```

```python
df_addr.withColumn('Addr_clean', clean_char(col('Addr'))).display()
```

```python
@udf
def remove_emoji(addr):
  return emoji.replace_emoji(addr,"")
```

```python
df_addr.withColumn('Addr_clean', remove_emoji(col('Addr'))).display()
```

```python
df_addr.withColumn('Addr', remove_space(col('Addr'))).display()
```

## From_json and to_json

```python
s1 = '''{'ZIP_CD':'212211','ADDRESS':'ASUGWMS','AREA_CD':32}'''
df_json = spark.createDataFrame([(12,s1),(14,s1)],schema=['ID','VALUE'])
df_json.display()
```

```python
json_schema = StructType([
    StructField('ZIP_CD' ,StringType(),True),
    StructField('ADDRESS' ,StringType(),True),
    StructField('AREA_CD' ,IntegerType(),True)
])
```

```python
df_json = df_json.withColumn('address',from_json(col('VALUE'),schema = json_schema))
df_json.display()
```

```python
df_json.select('address.*').display()
```

## to_json is used to convert struct type to string

```python
df_json.withColumn('addr', to_json(col("address"))).display()
```
