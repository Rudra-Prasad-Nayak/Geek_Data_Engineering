```python
df_test = spark.createDataFrame([('AB,CD',23,56),('NB,TY',43,78)],schema=['XY','MN','PQ'])
df_test.display()
```

```python
df_test = df_test.withColumn('XY_1', split(col('XY'),','))
df_test.display()
```

```python
df_test = df_test.withColumn('XY_2', explode(col('XY_1')))
df_test.display()
```

```python
df_test = df_test.withColumn('ABC',array('MN','PQ')).withColumn('XZ',arrays_zip('XY_1','ABC'))
df_test.display()
```

```python
df_test.select(explode('XZ').alias('XZ_1')).select('XZ_1.*').distinct().display()
```

```python
df_test.select(explode('XZ').alias('XZ_1')).select('XZ_1.XY_1','XZ_1.ABC').distinct().display()
```
