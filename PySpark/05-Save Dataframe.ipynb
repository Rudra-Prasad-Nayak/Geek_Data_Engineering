```python
df_flight = spark.read.format('csv').option('header', True).load('/Volumes/workspace/default/datasets/Flight_data.csv')
df_Indigo = df_flight.filter(lower(col('Airline'))=='indigo')
```

```python
Schema enforcement -  We are using a predefined schema while reading a file instead of inferschema.

schema Evolution - while writing a dataframe to a table, if we encounter any changes to column or datatypes, we are using Schema Evolution without loosing the historical data.
```

```python
schema evolution method --
Merge Schema -  when new column got added to the existing flow.we have to use schema Evolution
Overwrite Schema - when any column datatype got changed , we have to use overwrite schema.
```
