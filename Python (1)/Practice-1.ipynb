## prime number

```python
num = int(input('Enter a number:')) #5

c = 0
for i in range(1, num+1):
  if num%i == 0:
    c += 1

if c==2:
  print('prime number')
else:
  print('not a prime number')
```

## write a code to check whether a number is palindrome number

```python
num=121
if int(str(num)[::-1])==num:
    print('palindrome')
else:
    print('not a palindrome')
```

## how to find duplicates from a list

```python
ls3 = [4,3,6,2,3,5,4]
ls3
```

```python
unique_list = []
duplicate_list = []
for i in ls3:
    if ls3.count(i)>1:
        duplicate_list.append(i)
    else:
        unique_list.append(i)

unique_list, list(set(duplicate_list))
```
