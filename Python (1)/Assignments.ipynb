```python
# Wap to enter a five digit number and check whether it is palindrome or not.
num = int(input("Enter a 5 digit number"))

rev = int(str(num)[::-1])
print(rev)
if num == rev :
    print('Pallindrome')
else:
    print('Not Pallindrome')
```

```python
# Wap to enter a three digit number and check whether it is armstrong or not.
num = int(input("Enter a 5 digit number"))
sum = 0
len(str(num))

for i in str(num):
    sum = sum + int(i)**3

if sum == num:
    print('Armstrong Number')
else:
    print('Not Armstrong Number')
```

```python
# Wap to enter a character and check whether it is vowel or consonant by using match case statement.

char = input("Enter a character")

match char.upper():
    case 'A' | 'E' | 'I' | 'O' | 'U':
        print('Vowel')
    case _:
        print('Consonent')
```

```python
# print like this
#     *
#   * *
# * * *

n = int(input('Enter a number: '))

for i in range(1, n + 1):
    for j in range(n - i):
        print(' ', end='')
    for k in range(i):
        print('*', end='')
    print(end='\n')
```
