```python
class ABC:
    def __init__(this,num1,num2,num3):
        this.num1=num1
        this.num2=num2
        this.num3=num3
    
    def add(this):
        return this.num1+this.num2+this.num3
    
    def subtract(this,n):
        greatest_num = max(this.num1, this.num2, this.num3)
        return greatest_num-n
    
    def multiply(this):
        return this.num1*this.num2
    
    @staticmethod
    def div(num1,n):
        return num1%n

num1 = int(input('Enter a number1'))
num2 = int(input('Enter a number2'))
num3 = int(input('Enter a number3'))

A1 = ABC(num1,num2,num3)
print(A1.add())
print(A1.subtract(5))
print(A1.multiply())
print(A1.div(10, 4))
print(ABC.div(10, 4))
```

```python
class XYZ:
    percent = 10

    def __init__(this,num1,num2,num3):
        this.num1=num1
        this.num2=num2
        this.num3=num3
    
    def add(this):
        return this.num1+this.num2+this.num3
    
    @classmethod
    def update_precent(cls,n):
        cls.percent+=n
        return cls.percent
    
    def update_num(this):
        return (this.num1*XYZ.percent)//100 # XYZ.percent = 15 (Updated)
    
    @staticmethod
    def div(num1,n):
        return num1%n

A1 = XYZ(num1=85, num2=43, num3=23)

print(A1.add())
print(A1.update_precent(5))
print(A1.update_num())
print(A1.div(10, 4))
print(XYZ.div(10, 4))
```

#### Use Case
For example, You want to build one application for employee details.
Where you have employee class and functions like employee salary range, employee experience,employee_designation (‘system_engineer, IT Analyst, Team Lead, Senior Software Engineer), employee user id,employee mail id.

Pass the parameter as f_name, l_name, DOJ

```python
from datetime import datetime
```

```python
class Employee:
  domain = 'nielson.com'
  hike_percent = 10

  def __init__(self, f_name, l_name, DOJ, SAL):
    self.f_name = f_name
    self.l_name = l_name
    self.DOJ = DOJ
    self.SAL = SAL

  @classmethod
  def domain_nm(cls, domain_nm):
    cls.domain = domain_nm
    return cls.domain
  
  @classmethod
  def sal_hike(cls, hike):
    cls.hike_percent = cls.hike_percent+hike
    return cls.hike_percent
  
  def new_sal(self):
    return self.SAL+(self.SAL*(Employee.hike_percent/100))

  def emp_exp(self):
    return round((datetime.now() - datetime.strptime(self.DOJ, '%Y-%m-%d')).days / 365, 2)
  
  def emp_designation(self):
    exp_yrs = self.emp_exp()

    if exp_yrs > 0 and exp_yrs < 3:
      return 'SYSTEM ENGINEER'
    elif exp_yrs >= 3 and exp_yrs < 6:
      return 'IT ANALYST'
    elif exp_yrs >= 6 and exp_yrs < 10:
      return 'SENIOR SOFTWARE ENGINEER'
    elif exp_yrs >= 10 and exp_yrs < 15:
      return 'ASSISTANT CONSULTANT'
    else:
      return 'CONSULTANT'
  
  def emp_sal(self):
    emp_desig = self.emp_designation()

    if emp_desig == 'SYSTEM ENGINEER':
      return '3L-6L'
    if emp_desig == 'IT ANALYST':
      return '6L-9L'
    if emp_desig == 'SENIOR SOFTWARE ENGINEER':
      return '9L-12L'
    if emp_desig == 'ASSISTANT CONSULTANT':
      return '12L-16L'
    else:
      return '16L+'

  def user_id(self):
    # yr = datetime.strptime(self.DOJ, '%Y-%m-%d').year
    # mon = datetime.strptime(self.DOJ, '%Y-%m-%d').month
    return self.f_name[:2] + self.l_name[:2] + '_' + self.DOJ.split('-')[1] + self.DOJ.split('-')[0]
    # return self.f_name[:2] + self.l_name[:2] + '_' + str(mon) + str(yr)
  
  def email_id(self):
    email = self.f_name + '.' + self.l_name + str(datetime.strptime(self.DOJ, '%Y-%m-%d').year) + '@' + Employee.domain
    return email.lower()
```
